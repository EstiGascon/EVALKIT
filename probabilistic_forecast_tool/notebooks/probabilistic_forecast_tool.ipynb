# <span style="color: #4A90E2;">EvalKit - Probabilistic Forecast Analysis Tool</span>

## <span style="color: #4A90E2;">Overview</span>

EvalKit's Probabilistic Forecast Analysis Tool provides comprehensive ensemble forecast visualization and analysis capabilities. The application offers four distinct analysis types for examining forecast uncertainty and ensemble spread through interactive visualizations.

## <span style="color: #4A90E2;">Main Interface Components</span>

The application provides four plot type options:

1. **Meteogram Data** - Time series for single location with forecast range
2. **Plumes Data** - Ensemble plumes plots for uncertainty visualization
3. **Stamps Data** - Stamp plots for specific forecast date and step with ensemble spread
4. **CDF Data** - Cumulative Distribution Functions from climate data

---

## <span style="color: #4A90E2;">Step-by-Step Usage Guide</span>

### <span style="color: #5FD3A3;">Step 1: Select Plot Type</span>

Begin by choosing your desired analysis type from the four available options at the top of the interface. Each plot type offers unique visualization capabilities tailored to different analysis needs.

<img src="assets/img/plot_type_selection.png" alt="Plot Type Selection" width="100%" style="display: block; margin: 0 auto;">


---
### <span style="color: #5FD3A3;">Step 2: Configure Your Data Source</span>

After selecting a plot type, configure your data source. The interface adapts based on your plot type selection.

#### <span style="color: #7E8B99;">Option A: Download from MARS Archive</span>

##### **For Meteogram and Plumes Plot Types**

<img src="assets/img/meteogram_config.png" alt="Meteogram Configuration" width="100%" style="display: block; margin: 0 auto;">

1. **Data Source & MARS Configuration**:
   - **Model Class**: Select from dropdown (e.g., IFS - IFS)
   - **Forecast Initialization Date**: Select date (e.g., 01/02/2026)
   - **Forecast run time**: Choose from dropdown (e.g., 00:00:00)

2. **Grid Configuration**:
   - **Grid Resolution**: Enter grid resolution value (e.g., 0.25) for regular grid; if empty, reduced Gaussian grid is used

3. **Forecast Steps**:
   - **Forecast Steps (hours)**: Enter step interval (e.g., 0-24 for 24-hour forecast)
   - **Available Steps**: Review the generated forecast steps in the scrollable list
   - Use **"Select All"** or **"Deselect All"** buttons to manage step selection

4. **Parameter & Analysis Controls**:
   - **Parameter Selection**: Choose meteorological parameters from the scrollable list

5. **Geographic Area**:
   - **North**: Latitude (e.g., 72)
   - **South**: Latitude (e.g., 34)
   - **West**: Longitude (e.g., -25)
   - **East**: Longitude (e.g., 45)
      - Use the interactive map below to visually verify your area selection. You can either enter the coordinates directly into the four boxes, or draw a rectangle on the map to cover your desired area and the boxes will be automatically populated

6. **Action Buttons**:
   - **Validate Configuration**: Review your configuration before downloading
   - **Retrieve Data**: Start the data download

**Data Retrieved**: For Meteogram and Plumes, this retrieves both **perturbed forecasts** and **control forecast** for each selected parameter.

> **📝 Note**: When switching between Meteogram and Plumes plot types, your configuration and loaded data are preserved since both use the same input data structure.

---

##### **For Stamps Plot Type**

The configuration interface for Stamps is similar to Meteogram/Plumes, with the same sections for:
- Model Class selection
- Forecast initialization date and time
- Grid configuration
- Forecast steps
- Parameter selection
- Geographic area

**Data Retrieved**: For Stamps, the system retrieves:
- **Control forecast**
- **Perturbed forecasts** (ensemble members)
- **Deterministic forecast**

---

##### **For CDF Plot Type**

<img src="assets/img/cdf_config.png" alt="CDF Configuration" width="100%" style="display: block; margin: 0 auto;">

The CDF plot type has a different configuration structure optimized for climate analysis:

1. **Data Source & MARS Configuration**:
   - **Model Class**: Select from dropdown (e.g., IFS - IFS)
   - **Forecast Analysis Date**: Select date (e.g., 01/02/2026)

2. **CDF Analysis Settings**:
   - **Days Back**: Use the slider to specify the number of days to look back from the analysis date (e.g., 3 days)
   - **Forecast run time**: Select which forecast initialization times to include (e.g., Both (00Z & 12Z))

3. **Grid Configuration**:
   - **Grid Resolution**: Enter grid resolution value; if empty, reduced Gaussian grid is used

4. **Parameter Selection**:
   - Choose meteorological parameters for CDF analysis from the scrollable list

5. **Geographic Area**:
   - **North**: Latitude (e.g., 72)
   - **South**: Latitude (e.g., 34)
   - **West**: Longitude (e.g., -25)
   - **East**: Longitude (e.g., 45)
      - Use the interactive map below to visually verify your area selection. You can either enter the coordinates directly into the four boxes, or draw a rectangle on the map to cover your desired area and the boxes will be automatically populated

6. **Action Buttons**:
   - **Validate Configuration**: Review your configuration
   - **Retrieve Data**: Start the data download

**Data Retrieved**: The CDF analysis retrieves climate data and perturbed forecasts for each day and forecast run time within the specified lookback period for the selected parameters, enabling cumulative distribution function analysis.

---

#### <span style="color: #7E8B99;">Option B: Load Local Files</span>

For all plot types, you can load previously downloaded data files instead of retrieving from MARS Archive.

1. **Select Data Source**: Choose "Load Local Files" radio button

The file selection interface adapts based on your selected plot type:

##### **For Meteogram and Plumes Plot Types**

<img src="assets/img/plumes_local_config.png" alt="Load Local Files - Meteogram/Plumes" width="100%" style="display: block; margin: 0 auto;">

Both plot types require two forecast files:

2. **File Selection**:
   - **Ensemble Forecast File Path**: 
     - Enter the file path directly in the text box, or
     - Click **"Browse"** button to select your perturbed ensemble forecast file
   - **Control Forecast File Path**:
     - Enter the file path directly in the text box, or
     - Click **"Browse"** button to select your control forecast file

3. **Action Buttons**:
   - Click **"Validate Configuration"** to verify your files
   - Click **"Load Data"** to load the selected files

> **📝 Note**: Data and configuration are preserved when switching between Meteogram and Plumes plot types since both use the same input file structure (ensemble and control forecasts).

---

##### **For Stamps Plot Type**

<img src="assets/img/stamps_local_config.png" alt="Load Local Files - Stamps" width="100%" style="display: block; margin: 0 auto;">

Stamps visualization requires three forecast files:

2. **File Selection**:
   - **Deterministic Forecast File Path**:
     - Enter the file path directly in the text box, or
     - Click **"Browse"** button to select your deterministic forecast file
   - **Control Forecast File Path**:
     - Enter the file path directly in the text box, or
     - Click **"Browse"** button to select your control forecast file
   - **Ensemble Forecast File Path**:
     - Enter the file path directly in the text box, or
     - Click **"Browse"** button to select your perturbed ensemble forecast file

3. **Action Buttons**:
   - Click **"Validate Configuration"** to verify your files
   - Click **"Load Data"** to load the selected files

---

##### **For CDF Plot Type**

<img src="assets/img/cdf_local_config.png" alt="Load Local Files - CDF" width="100%" style="display: block; margin: 0 auto;">

CDF analysis requires climate reference data and multiple scenario files for different days and forecast times:

2. **Climate Data File**:
   - **Climate Data File Path**:
     - Enter the file path directly in the text box, or
     - Click **"Browse"** button to select your climate data file

3. **Scenario Files Configuration**:
   
   For each scenario, you can configure:
   - **Scenario File**: Click **"Browse"** to select the file or enter the file path directly in the text box (shows "No file selected" until chosen)
   - **Days Back**: Select how many days back from the analysis date (dropdown: 0, 1, 2, 3...)
   - **Time**: Select the forecast initialization time (dropdown: 00Z, 12Z)
   - **Browse**: Opens file browser to select the scenario file
   - **Delete**: Removes this scenario row

   The interface provides multiple scenario rows (e.g., D-0_00Z, D-0_12Z, D-1_00Z, D-1_12Z, D-2_00Z, etc.) where:
   - **D-0**: Analysis day (today)
   - **D-1**: 1 day back
   - **D-2**: 2 days back
   - **D-3**: 3 days back

4. **Managing Scenarios**:
   - Click **"Add More Scenario Files"** to add additional scenario configurations
   - Use **"Delete"** button to remove unwanted scenario rows
   - Configure as many scenarios as needed for your analysis

5. **Action Buttons**:
   - Click **"Validate Configuration"** to verify all files and settings
   - Click **"Load Data"** to load the climate data and all scenario files

### <span style="color: #5FD3A3;">Step 3: Observation Data Integration (Optional , Meteogram & Plumes Only)</span>

<img src="assets/img/obs_integration.png" alt="Observation Integration - No" width="100%" style="display: block; margin: 0 auto;">

For Meteogram and Plumes plot types, you have the option to integrate observation data for model-observation comparison.

1. **Include Observation Data?**
   - Select **"No"** if you only want to visualize forecast data
   - Select **"Yes"** to enable observation data integration

<img src="assets/img/obs_existing_folder.png" alt="Observation Integration - Yes" width="100%" style="display: block; margin: 0 auto;">

#### **When "Yes" is selected, two data source options become available:**

##### **Option A: Browse Existing Folder**

Use this option if you already have observation data stored locally.

1. Select **"Browse existing folder"** radio button
2. **Observation Folder**: 
   - The folder path selection field will appear with "No folder selected"
   - Click **"Browse Folder"** button to navigate to your observation data directory
   - Select the folder containing your observation files

<img src="assets/img/obs_retrieval_config.png" alt="Observation Integration - Retrieve New" width="100%" style="display: block; margin: 0 auto;">

##### **Option B: Retrieve New Observations**

Use this option to download fresh observation data using STVL.

1. Select **"Retrieve new observations"** radio button

2. **Retrieve Configuration**:
   - **Parameter to Retrieve**: Select the meteorological parameter from dropdown (e.g., 2m Temperature)
   - **Data Sources**: Choose observation sources from dropdown (e.g., SYNOP, HDOBS or Both)
   - **Period**: Select the accumulation period for period-based parameters (e.g., 6 hours, 12 hours, 24 hours)
     - *Note: This field is only applicable for parameters like precipitation, wind gust, and temperature extremes*
   - **Retrieval Times (UTC)**: Displays which UTC hours will be retrieved based on your parameter and period selection
     - Times are automatically determined based on the parameter type and period
     - For instantaneous parameters: observations every 3 hours (0, 3, 6, 9, 12, 15, 18, 21 UTC)
     - For period-based parameters: times vary with the selected period

3. **Date Range & Output**:
   - **Start Date**: Enter start date in format `DD/MM/YYYY` (e.g., 30/01/2026)
   - **End Date**: Enter end date in format `DD/MM/YYYY` (e.g., 01/02/2026)
   - **Output Directory**: Specify where to save retrieved observations (e.g., `/retrieved_observations`)
   - Click **"Browse"** to select a different output directory if needed

4. **Retrieve Observations**:
   - Click the **"Retrieve Observations"** button to download the observation data
   - The system will fetch observation data for the specified date range and selected parameter
   - Retrieved data will be automatically saved to the specified output directory

> **📝 Note**: Observation data integration is only available for Meteogram and Plumes plot types.

### <span style="color: #5FD3A3;">Step 4: Automatic Visualization Platform</span>

After loading your data, the Automatic Visualization Platform enables quick plot generation with minimal configuration.

---

#### <span style="color: #7E8B99;">For Meteogram and Plumes Plot Types</span>

<img src="assets/img/meteogram_plotting_config.png" alt="Meteogram Plotting Configuration" width="100%" style="display: block; margin: 0 auto;">

1. **Plotting Configuration** (Left Panel):
   - **Parameter Selection**: Choose a meteorological parameter from the dropdown menu
   - Available parameters are displayed in a scrollable list (e.g., 10 metre U wind component, 10 metre V wind component, 2 metre temperature, 6-hour precipitation, etc.)
   - **Unit Selection**: Select your preferred unit from the dropdown (e.g., mm for precipitation, °C for temperature)
   - **Action Buttons**:
     - **Clear Plot**: Remove the current visualization
     - **Refresh Plot**: Update the plot with new settings

2. **Geographic Selection** (Right Panel):
   - **Interactive Map**: Click on the map to select ONE analysis point for meteogram plots
   - The point must be within the defined geographic area
   - A blue marker will appear at your selected location
   - Clicking a new location will move the analysis point

   **Alternative Manual Entry**:
   - **Lat**: Enter latitude coordinate manually
   - **Lon**: Enter longitude coordinate manually
   - Click **"+ Add Point"** to place the marker
   - Click **"Clear Point"** to remove the current point

3. **Plot Generation**:
   - Once a point is selected, the system automatically generates the plot
   - A status message appears: "Creating meteogram plot automatically..."
   - When complete: "Success - Meteogram created! Parameter: [parameter name]"

<img src="assets/img/meteogram_output.png" alt="Meteogram Plot Output" width="100%" style="display: block; margin: 0 auto;">

**Meteogram Plot Features**:
- **Red line**: Control Forecast
- **Blue boxes**: Ensemble Members (showing spread/uncertainty)
- Time series visualization with forecast range
- Interactive toolbar for zoom, pan, and export options


**Plumes Plot Features**:
- **Orange line**: Control Forecast
- **Shaded green areas**: Ensemble spread ranges
  - Light green: 0%-100% range
  - Medium green: 5%-95% range
  - Dark green: 25%-75% range (IQR - Interquartile Range)
  - Darkest green line: Median (50%)
- Time series showing forecast uncertainty evolution

> **📝 Note**: Configuration and data are preserved when switching between Meteogram and Plumes plot types. Simply select the different plot type button to switch visualization styles for the same location and parameter.
