# <span style="color: #171A35;">EvalKit - Interactive Weather Model Timeseries Analysis</span>

**EvalKit** is a comprehensive weather model analysis tool designed for meteorological data evaluation and visualization.

## <span style="color: #00BCD4;">Interface Overview</span>

The application consists of two main panels:

### <span style="color: #50DEA3;">Configuration Panel</span>
Controls for data setup and analysis parameters, organized into three sections:
- **Data Source** (Left): Configure data retrieval or local file loading
- **Parameter & Analysis** (Center): Select parameters and models to analyze
- **Observation Data** (Right): Configure optional observation data integration

<img src="assets/img/data_acq.png" alt="Visualization Panel" width="100%" style="display: block; margin: 0 auto;">

### <span style="color: #50DEA3;">Visualization Panel</span>
- **Left Pane**: Interactive time series chart with dynamic legend and analysis tools
- **Right Pane**: Geographic map with forecast points, observation stations, and drawing controls

## <span style="color: #00BCD4;">How to Use EvalKit</span>

### <span style="color: #616161;">Step 1: Get Forecast Data</span>

You have two options for obtaining forecast data:

**Option A: Download from MARS Archive**
1. Select "Download from MARS Archive"
2. Choose your model from the dropdown
3. Set the time period using date pickers and time selection
4. Define the geographic area by drawing a bounding box on the map or entering coordinates manually
5. Set grid resolution if needed
6. Click "Preview" to review settings, then "Retrieve Data"

**Option B: Load Local Files**
1. Select "Load Local File(s)"
2. Click "Browse AIFS" and/or "Browse IFS" to select your model files, or add the paths of your grib files.
3. Click "Load File" to process the selected files
4. Verify successful loading in the Loading Summary panel

### <span style="color: #616161;">Step 2: Select Analysis Parameter</span>

1. Choose the meteorological parameter you want to analyze from the dropdown menu
2. Check the boxes for the models you want to compare (AIFS Model, IFS Model)

### <span style="color: #616161;">Step 3: Add Observation Data (Optional)</span>

If you want to compare forecast data with observations:

**If you have existing observation data:**
1. Select "Yes" for "Do you have observation data?"
2. Select "Browse existing folder" or add its path.
3. Click "Browse Observations" to select your observation data folder
4. The system will automatically validate parameter compatibility

**If you need to retrieve observation data:**
1. Select "Yes" for "Do you have observation data?"
2. Select "Retrieve new observations"
3. Configure the VINO path to your vino_getgeo executable
4. Choose data sources (SYNOP, HDOBS, or Both)
5. Set observation period and time range
6. Choose output folder and click "Retrieve Observations"

Once observation data is loaded, check the "Observations" box to include it in your analysis.

### <span style="color: #616161;">Step 4: Select Analysis Points</span>

1. Click on locations within your defined geographic area on the map or add cooridnates manually.
2. Selected points appear as colored markers
3. Select multiple points to compare different locations
4. Orange markers represent observation stations (if loaded) - click to select them

### <span style="color: #616161;">Step 5: Analyze Results</span>

<img src="assets/img/viz_section.png" alt="Visualization Panel" width="100%" style="display: block; margin: 0 auto;">

1. View the time series chart showing data for all selected points
2. Each point/station has a unique color matching its map marker
3. Use chart tools for zooming and detailed inspection
4. Compare forecast models with observations

### <span style="color: #616161;">Step 6: Manage Your Analysis</span>

- **Clear All Points**: Remove all selected forecast points and observation stations
- **Clear Drawings**: Remove bounding boxes from the map
- **Modify Selection**: Click existing points to remove them, or add new points

## <span style="color: #00BCD4;">Tips for Best Results</span>

- Start with smaller geographic areas and shorter time periods for initial analysis
- Always preview your settings before retrieving large datasets from MARS
- Draw bounding boxes on the map for precise geographic area selection
- When using observations, ensure the time period matches your forecast data
- The system will warn you if observation and forecast parameters don't match

## <span style="color: #00BCD4;">Typical Workflow</span>

1. **Get Data**: Download from MARS or load local files
2. **Choose Parameter**: Select what you want to analyze
3. **Add Observations**: Load observation data if available
4. **Select Points**: Click on map to choose analysis locations  
5. **Compare**: View time series plots of forecasts vs observations
