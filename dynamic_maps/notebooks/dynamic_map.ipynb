# <span style="color: #4A90E2;">EvalKit - Interactive Surface Variables Visualization Tool</span>

## <span style="color: #4A90E2;">Overview</span>

EvalKit is a comprehensive meteorological analysis tool designed for visualizing and analyzing surface weather variables. The application provides an interactive interface for downloading, processing, and visualizing weather model data with integrated mapping capabilities.

## <span style="color: #4A90E2;">Main Interface Components</span>

The application consists of three main collapsible panels:

1. **Configuration Panel** - Set up data sources and analysis parameters
2. **Surface Variables Calculator** - Compute derived meteorological variables
3. **Interactive Plotting** - Create and customize visualizations

---

## <span style="color: #4A90E2;">Step-by-Step Usage Guide</span>

### <span style="color: #5FD3A3;">Step 1: Configure Your Data Source</span>

The Configuration Panel offers two data source options:

#### <span style="color: #7E8B99;">Option A: Download from MARS Archive</span>

1. **Select Data Source**: Choose "Download from MARS Archive" radio button

2. **Parameter Selection**: 
   - Select meteorological parameters from the scrollable list (e.g., 2m Temperature, Wind components, Precipitation).
   - Multiple parameters can be selected

3. **Model & Time Settings**:
   - **Model**: Select from dropdown (e.g., IFS Operational)
   - **Date**: Enter date in format `YYYY-MM-DD` (e.g., 2026-03-29)
   - **Time**: Choose from dropdown (e.g., 00:00:00)

4. **Forecast Steps**:
   - **Start Step**: Enter starting forecast hour (e.g., 0)
   - **End Step**: Enter ending forecast hour (e.g., 72)
   - Click **"Update Steps"** button to apply
   - Use **"Select All Steps"** or **"Deselect All"** buttons as needed
   - Review selected steps in the scrollable list below

5. **Geographic Area**:
   - **North**: Latitude (e.g., 72)
   - **South**: Latitude (e.g., 34)
   - **West**: Longitude (e.g., -25)
   - **East**: Longitude (e.g., 45)
   - Use the interactive map below to visually verify your area selection. You can either enter the coordinates directly into the four boxes, or draw a rectangle on the map to cover your desired area and the boxes will be automatically populated

6. **Grid Resolution**:
   - Enter grid resolution value (e.g., 0.25) for regular grid,  if empty reduced gaussian grid is used.
   - Click **"Reset Grid"** or **"Clear Grid"** as needed

7. **Action Buttons**:
   - **Preview Settings**: Review your configuration before downloading
   - **Retrieve Data**: Start the data download
   - **Reset All**: Clear all settings

#### <span style="color: #7E8B99;">Option B: Load Local File</span>

1. **Select Data Source**: Choose "Load Local File" radio button

2. **File Selection**:
   - **File Path**: Enter the path to your file directly in the box (e.g., `path/to/yourdata.grib`), or click **"Browse Files"** to select a file using the file browser
   - Once your file is selected, click **"Load File"** to load it

3. **Interactive Map Selection**:
   - The map shows your geographic coverage
   - Use zoom controls (+/-) to navigate
   - Use the drawing tool (square icon) to select specific regions

### <span style="color: #5FD3A3;">Step 2: Calculate Derived Variables (Optional)</span>

If you need to compute derived meteorological variables:

1. **Access Calculator**: Expand the "Surface Variables Calculator" panel

2. **Select Calculation Type**:
   - Choose from the dropdown menu (e.g., Wind Speed, Accumulated Precipitation, Extremes)
   - Based on the available data you've loaded, you will receive an informative message indicating whether this type of calculation is possible for your dataset or not
   - The calculator will use data from the variables you've loaded

3. **Execute Calculation**:
   - Click the **"Calculate"** button
   - Computed variables will be added to your available parameters

<img src="assets/img/calculator_sv.png" alt="Visualization Panel" width="100%" style="display: block; margin: 0 auto;">

   **Available Calculation Methods**

1. **Wind Speed Calculation** 
- **Inputs**: 10m U (`u10`) and V (`v10`) wind components  
- **Output**: Instantaneous wind speed magnitude at each timestep  

2. **Accumulated Precipitation**   
- **Input**: Total Precipitation  (`tp`)  
- **Output**: Total precipitation accumulation between first and last forecast step  

3. **Max/Min Values** 
- **Input**: Any time-varying variable (temperature, wind-gust, etc.)  
- **Outputs**:  
  - **Maximum**: Highest value at each grid point across all timesteps  
  - **Minimum**: Lowest value at each grid point across all timesteps

## <span style="color: #4A90E2;">Customization and Extensibility</span>

### <span style="color: #5FD3A3;">Adding New Models and Parameters</span>

EvalKit is designed to be **scalable and customizable**. You can extend the tool to work with additional weather models and parameters by modifying the configuration files.

#### <span style="color: #7E8B99;">Model Configuration</span>

To add new weather models to the system, you can modify the model configuration settings in `helpers/config.json`. The system supports models from MARS Archive with different configurations including:

- **Model specifications**: class, stream, type, and level type
- **Step patterns**: Define forecast time intervals (fixed or date-based)
- **Custom step expansion**: Control whether forecast steps can be customized

**Example Model Configuration:**
```json
"new-model": {
  "display_name": "New Model Name",
  "class": "od",
  "stream": "oper",
  "type": "fc",
  "levtype": "sfc",
  "step_pattern": "six_hourly",
  "supports_custom_step_expansion": false,
  "description": "Description of the new model",
  "ui_order": 3
}
```

#### <span style="color: #7E8B99;">Parameter Configuration</span>

New meteorological parameters can be added to the system. When adding parameters, ensure they are compatible with the **same step configuration** as existing parameters in your model.

> **⚠️ Important Note on Step Compatibility:**
> 
> Parameters with different temporal resolutions (e.g., 6-hourly vs. 1-hourly data) should not be mixed in the same analysis. For example, the --hour 10m Wind Gust parameter (`10fg6`) typically has 6-hour intervals, while other parameters like 2m Temperature may have 1-hour intervals. Adding parameters with incompatible step configurations may cause data alignment issues.

**Example Parameter Configuration:**
```json
"new_param": {
  "param_id": 999,
  "name": "New Parameter Name",
  "display_name": "Display Name",
  "units": "unit",
  "description": "Parameter description",
  "available_models": ["model-name"],
  "category": "category_name",
  "ui_order": 10
}
```

#### <span style="color: #7E8B99;">Visualization Styling</span>

To customize the visualization appearance of new parameters, you need to modify the **`styling_config.py`** file located in the `helpers/` directory. This file controls:

- **Color palettes**: RGB color schemes for different parameter types
- **Level definitions**: Value ranges for contour levels
- **Unit conversions**: Transformations between different measurement units
- **Parameter categorization**: Grouping parameters by type (temperature, precipitation, wind, etc.)

**Steps to Add New Parameter Styling:**

1. Open `helpers/styling_config.py`
2. Add your parameter to the appropriate category in `parameter_types` dictionary
3. Define color palette and levels if using a new parameter type
4. Add title formatting in the appropriate `_get_*_title()` method
5. If needed, add unit conversion logic in `transform_data_and_levels()` method

**Example - Adding a new parameter category:**
```python
self.parameter_types = {
    # Existing categories...
    "new_category": ["new_param1", "new_param2"],
}

# Define colors and levels for the new category
self.new_category_colors = [
    "rgb(0,0,255)",
    "rgb(0,255,0)",
    # ... more colors
]
self.new_category_levels = [0, 10, 20, 30, 40, 50]
```
